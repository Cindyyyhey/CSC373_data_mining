````python
def scaling(data):
    # Drop the response variable class since we only need to scale numeric variables
    # Store the new dataset to a dataframe called df
    if 'Class' in data.columns:
        data_scale = data.drop('Class', 1)
    else:
        data_scale = data

    # Loop through each column and scale the values
    # write the scaled value back to df
    for col in list(data_scale.columns):
        data_scale[col] = (data_scale[[col]]-data_scale[[col]].min())/(data_scale[[col]].max()-data_scale[[col]].min())
  
    # print out the scaled dataset
    display(data_scale)
    return(data_scale)
````

# call the scaling function and add the class back to our dataset

````python
data_scale = scaling(data)
data_scale["Class"] = data["Class"]
display(data_scale)
````

````python
mean_dic = {}
for col in list(data_scale.columns):
    if col!="Class":
        #print(col,np.mean(data_scale[col]))
        mean_dic[col] = np.mean(data_scale[col])
````

# The function for calculating entropy

````python
def entropyCal(count1,count2):
    totalC = count1 + count2
    return -(count1/totalC)*np.log2(count1/totalC) -(count2/totalC)*np.log2(count2/totalC)
````

````python
entropy_root = entropyCal(data_scale['Class'].value_counts()["B"],data_scale['Class'].value_counts()["M"])
entropy_root
````

# The function for calculating gini

````python
def giniCal(count1,count2):
    totalC = count1 + count2
    return 1-(pow((count1/totalC),2)+pow((count2/totalC),2))
````

````python
gini_root = giniCal(data_scale['Class'].value_counts()["B"],data_scale['Class'].value_counts()["M"])
gini_root
````

````python
for col in list(data_scale.columns):
    if col!="Class":
        child1_count = sum(data_scale[col] < mean_dic[col])
        child2_count = sum(data_scale[col] >= mean_dic[col])
        child_total = child1_count + child2_count
        
        child1_count_B = data_scale[data_scale[col] < mean_dic[col]]["Class"].value_counts()["B"]
        child1_count_M = data_scale[data_scale[col] < mean_dic[col]]["Class"].value_counts()["M"]
        
        gini_child1 = giniCal(child1_count_B,child1_count_M)
        entropy_child1 = entropyCal(child1_count_B,child1_count_M)
        
        child2_count_B = data_scale[data_scale[col] >= mean_dic[col]]["Class"].value_counts()["B"]
        child2_count_M = data_scale[data_scale[col] >= mean_dic[col]]["Class"].value_counts()["M"]
        
        gini_child2 = giniCal(child2_count_B,child2_count_M)
        entropy_child2 = entropyCal(child2_count_B,child2_count_M)
        
        gini_commbined = child1_count/child_total*gini_child1 + child2_count/child_total*gini_child2
        entropy_commbined = child1_count/child_total*entropy_child1 + child2_count/child_total*entropy_child2
        print(gini_commbined,entropy_commbined)
        
````
